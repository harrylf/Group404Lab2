```python
import pyaudio
import socket


CHUNK = 1024
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 44100

p = pyaudio.PyAudio()

stream = p.open(format = FORMAT,
                channels = CHANNELS,
                rate = RATE,
                input = True,
                frames_per_buffer = CHUNK,
                )

udp = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)

while True:
    udp.sendto(stream.read(CHUNK), ("127.0.0.1", 4444))
    print('I am here')

udp.close()
```
